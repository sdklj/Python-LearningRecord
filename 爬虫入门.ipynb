```python
coding：utf-8
# 引入相关模块
import requests
from bs4 import BeautifulSoup

url = "http://www.dilidili1.com/"
```

```python

#请求嘀哩嘀哩的URL，获取其text文本
webRaw = requests.get(url)
webRaw.encoding = 'utf8' 
wbdata = webRaw.text
#print(wbdata)
#对获取到的文本进行解析
soup = BeautifulSoup(wbdata,'lxml')
# 从解析文件中通过select 选择器定位指定的元素，返回一个列表
new_titles = soup.select(".am-list-item-hd")#参数选择有问题
#print(new_titles)

#对返回的列表进行遍历
for n in new_titles:
    #提取出标题和链接信息
    title = n.get_text()
    link = n.get("herf")
    data = {
        '标题':title,
        '链接':link
    }
    print(data)
```
